```python
mean = 146.41216216216216
# we replace the unknown loan amount with the mean
data["LoanAmount"].fillna( mean, inplace=True)
data.isnull().sum()
```

```python
# data visualization
# a simple pie chart male v fermale
# we are going to use matplotlib for visualisation
import matplotlib.pyplot as plt

# get the count of male vs female
genderCount= data['Gender'].value_counts()

# define the size of the figure the width and the height in inches
plt.figure(figsize=(6,6))
plt.pie(genderCount,labels=genderCount.index, autopct='%1.1f%%',colors= ['blue', 'pink'], startangle=90 )

plt.title = ('Gender distribution')
plt.show()
```

```python
# bar plot
import matplotlib as plt
import seaborn as sns
import numpy as np

# we want to plot a barplot using seabobron
sns.barplot(data =data , x="Gender", y="LoanAmount", estimator=np.mean , hue="Married" ,palette="Blues")

plt.title(" Barplot of gender v loan amount")
plt.xlabel("Gender")
plt.ylabel("LOanAmount")
plt.savefig ('gender_distribution.pdf')
```
