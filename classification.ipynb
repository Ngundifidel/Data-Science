```python
import pandas
data = pandas.read_csv("data/data.csv")
data
```

```python
# check if we have any empties
data.isnull().sum()
```

```python
data.groupby("Type").size()
```

```python
# split the data into different data setsx 
# x data set (Quantitative data type)
# y data set (Qualitative data type)

array = data.values
array
X = array[:, 0:8]
# X
Y=  array[:, 8:9]
Y
```

```python
# when uyou have qualitative abd quantitaitve data, how do you do analysis with it. using classification
# import skelarn
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y,train_size=0.3, random_state=42)
```

```python
# lets pass our data through support vector classifier(SV)
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,Y_train )
```

```python
# lets test the model predication
svs_predict = model.predict(X_test)
svs_predict
```

```python
# check the mode accuracy score
from sklearn.metrics import accuracy_score
check_accuracy = accuracy_score(svs_predict , Y_test) * 100
check_accuracy 
```

```python
# Gradientboosting classifier
from sklearn.ensemble import GradientBoostingClassifier as gbc
model1 = gbc()
model1.fit(X_train,Y_train)


```

```python
gbc_predict = model1.predict(X_test)
gbc_predict
```

```python
from sklearn.metrics import accuracy_score
check_accuracy1 = accuracy_score(gbc_predict , Y_test) * 100
check_accuracy1
```

```python
from sklearn.linear_model import LogisticRegression as lr 
model2 = lr()
model2.fit(X_train,Y_train)

```

```python
lr_predict = model2.predict(X_test)
lr_predict
```

```python
from sklearn.metrics import accuracy_score
check_accuracy2 = accuracy_score(lr_predict , Y_test) * 100
check_accuracy2
```
