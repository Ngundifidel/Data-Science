```python
import pandas 
data = pandas.read_csv("data/Advertising.csv")
data
```

```python
# split the data in to qualitative and quantitative
array = data.values
array
X = array[:, 1: 4]

# X
Y=  array[:, 4:5]
# Y
```

```python
# when uyou have qualitative abd quantitaitve data, how do you do analysis with it. using classification
# import skelarn
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y,train_size=0.3, random_state=42)
```

```python
# lets pass our data through support vector classifier(SVr)
from sklearn.svm import SVR
model = SVR()
model.fit(X_train,Y_train )
```

```python
# lets test the model predication
svr_predict = model.predict(X_test)
svr_predict
```

```python
# check the mode accuracy score
from sklearn.metrics import r2_score
check_accuracy = r2_score(svr_predict , Y_test) * 100
check_accuracy 
```

```python
from sklearn.ensemble import RandomForestRegressor as rf 
model = rf()
model.fit(X_train,Y_train )
rf_predict = model.predict(X_test)
rf_predict
check_accuracy1 = r2_score(rf_predict , Y_test) * 100
check_accuracy1
```

```python
from sklearn.tree import DecisionTreeRegressor as dt 
model1 = dt()
model1.fit(X_train,Y_train )
dt_predict= model1.predict(X_test)
dt_predict
check_accuracy1 = r2_score(dt_predict , Y_test) * 100
check_accuracy1
```

```python
from sklearn.ensemble import GradientBoostingRegressor as gb 
model2 = gb()
model2.fit(X_train,Y_train )
gb_predict= model2.predict(X_test)
gb_predict
check_accuracy2 = r2_score(gb_predict , Y_test) * 100
check_accuracy2
```

```python
sales = [ [ 10,80, 10]]
svrcoutcome = model.predict(sales)
gbcoutcome = model1.predict(sales)
gradientoutcome = model2.predict(sales)

print (" svr says you sales will be", svrcoutcome)
print (" Decison treesays y you sales will be", gbcoutcome)
print (" Gradient boost says  you sales will be", gradientoutcome)
```
